```python
#Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:
"""
если источник traffic_source равен yandex или google, то в source_type ставится organic
для источников paid и email из России - ставим ad
для источников paid и email не из России - ставим other
все остальные варианты берем из traffic_source без изменений
"""

import pandas as pd

#чтение файлов
data_log = pd.read_csv(r'C:\Users\mazar\Documents\Python\Нетология - курс\Homework Unit13- Pandas2\visit_log.csv', sep=';')
```

```python
data_log.head()
```

```python
def source_type_fnc(row):
    source_type = row['traffic_source']

    region = row['region'].lower()
    traffic_source = row['traffic_source'].lower()
    
    if traffic_source in ['yandex', 'google']:
        source_type = 'organic'
    elif traffic_source in ['paid', 'email'] and region == 'russia':
        source_type = 'ad'        
    elif traffic_source in ['paid', 'email'] and region != 'russia':
        source_type = 'other'       
    return source_type
```

```python
#Задание 3
"""
Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, 
которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным 
значением столбца timestamp для данного значения userId.
"""
#чтение файлов
data_ratings = pd.read_csv(r'C:\Users\mazar\Documents\Python\Нетология - курс\ml-latest-small\ratings.csv')
data_ratings.head()
```

```python
len(data_ratings)
```

```python
#считаем кол-во записей по каждому пользователю и пишем в новый датафрейм
df = data_ratings.groupby('userId').agg({'movieId':'count'}).reset_index()
```

```python
#собираем список userId пользователей с >100 записей
df = df[df['movieId']>100]['userId']
```

```python
#оставляем оценки только пользователей из списка
df2 = data_ratings[data_ratings['userId'].isin(df)]
```

```python
df2.head()
```

```python
#по каждому пользователю считаем мин и макс дату. пишем в новый датафрейм
df3 = df2.groupby('userId').agg({'timestamp': ['min', 'max']}).reset_index()
```

```python
df3.head()
```

```python
#создаём новый столбец = разница между мин и макс
df3['diff'] = df3['timestamp']['max'] - df3['timestamp']['min']
```

```python
df3.head()
```

```python
#считаем среднюю разницу по всем подьзователям
df3['diff'].mean()
```

```python
#Задание 4
"""
Дана статистика услуг перевозок клиентов компании по типам (см. файл с кодом занятия). Необходимо сформировать две таблицы:
таблицу с тремя типами выручки для каждого client_id без указания адреса клиента
аналогичную таблицу по типам выручки с указанием адреса клиента
"""
```

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

```python
#с указанием адреса клиента
client_base.merge(rzd, how='left', on='client_id').merge(auto, how='left', on='client_id').merge(air, how='left', on='client_id')
```

```python
#без указания адреса клиента
rzd.merge(auto, how='outer', on='client_id').merge(air, how='outer', on='client_id')
```
